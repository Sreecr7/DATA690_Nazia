# Goods exports

```python
df_Export = df.query('indicator == "BX.GSR.MRCH.CD"')
df_Export.head(10)
```

```python
df_pivot = df_Export.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'],columns = 'indicator', values ='value').reset_index()
df_pivot.reset_index().head()
```

```python
fig = px.pie(df_pivot, values='BX.GSR.MRCH.CD', names='Country Name', title='Total Population',hover_name='Country Name')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
df_pivot = df_2010.pivot(index =['Country Code','Country Name','Region','Income Group','Lending Type'],columns = 'indicator', values ='value').reset_index()
df_pivot.reset_index().head()
```

#Imports of goods and services 

```python
df_Import = df.query('indicator == "BM.GSR.GNFS.CD"')
df_Import.head(10)
```

```python
df_pivot1 = df_Import.pivot(index =['Year','Country Code','Country Name','Region','Income Group','Lending Type'],columns = 'indicator', values ='value').reset_index()
df_pivot1.reset_index().head()
```

```python
fig = px.line(
    df_pivot1,
    x="Year", 
    y="BM.GSR.GNFS.CD",
    color='Country Name',
    hover_name = "Country Name",
    title='TOP 5 Countries by 2010 Imports of goods and services (BoP, current US$)',
    labels={"BM.GSR.GNFS.CD":"Imports of goods and services (BoP, current US$)"}
    )

fig.show()

```

##### Between 2010 and 2020, the United States' imports of goods and services fluctuated significantly. As a result, this could be a deciding factor in determining the economy.
